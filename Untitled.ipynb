# Geometrik şekilleri üç boyutlu olrak oluşturma

- Bu çalışma, 3D grafikleri kullanarak geometrik şekillerin görselleştirilmesine yönelik temel bir örnek sunmaktadır. Şekillerin merkezde konumlandırılması, görselliği artırır ve analiz yapmayı kolaylaştırır. Kodlar, matplotlib ve mpl_toolkits gibi kütüphaneleri kullanarak 3D görselleştirme sağlar ve her bir şeklin nasıl çizildiğini detaylı bir şekilde gösterir.

- Bu açıklama, geometrik şekillerin 3D ortamda nasıl çizildiğini ve her bir şeklin nasıl ortalandığını anlatmaktadır. Kodlar, her bir şeklin çizim sürecini adım adım açıklayacak şekilde düzenlenmiştir.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from mpl_toolkits.mplot3d import Axes3D
```

```python
def draw_centered_small_cube(ax):
    # Küpün köşeleri
    vertices = np.array([[-0.25, -0.25, -0.25], [0.25, -0.25, -0.25], 
                         [0.25, 0.25, -0.25], [-0.25, 0.25, -0.25],
                         [-0.25, -0.25, 0.25], [0.25, -0.25, 0.25], 
                         [0.25, 0.25, 0.25], [-0.25, 0.25, 0.25]])

    # Küpün yüzeyleri
    faces = [[vertices[j] for j in [0, 1, 2, 3]], [vertices[j] for j in [4, 5, 6, 7]],
             [vertices[j] for j in [0, 1, 5, 4]], [vertices[j] for j in [1, 2, 6, 5]],
             [vertices[j] for j in [2, 3, 7, 6]], [vertices[j] for j in [3, 0, 4, 7]]]

    ax.add_collection3d(Poly3DCollection(faces, facecolors='cyan', linewidths=1, edgecolors='r', alpha=.5))

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
draw_centered_small_cube(ax)
ax.set_title('Küp')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
plt.show()
```

```python
def draw_centered_small_pyramid(ax):
    base = np.array([
        [-0.25, -0.25, 0], [0.25, -0.25, 0], [0.25, 0.25, 0], [-0.25, 0.25, 0],  # Dört köşe
        [0, 0, 0.5]  # Apex
    ])
    
    faces = [
        [base[j] for j in [0, 1, 4]], [base[j] for j in [1, 2, 4]],
        [base[j] for j in [2, 3, 4]], [base[j] for j in [3, 0, 4]],
        [base[j] for j in [0, 1, 2, 3]]  # Taban yüzeyi
    ]
    
    ax.add_collection3d(Poly3DCollection(faces, facecolors='purple', linewidths=1, edgecolors='r', alpha=.5))

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
draw_centered_small_pyramid(ax)
ax.set_title('Piramit')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
plt.show()
```

```python
def draw_centered_small_triangular_prism(ax):
    base = np.array([
        [-0.25, -0.25, 0], [0.25, -0.25, 0], [0, np.sqrt(3)/8, 0],  # Taban köşeleri
        [-0.25, -0.25, 0.5], [0.25, -0.25, 0.5], [0, np.sqrt(3)/8, 0.5]  # Üst köşeler
    ])
    
    faces = [
        [base[j] for j in [0, 1, 2]], [base[j] for j in [3, 4, 5]],
        [base[j] for j in [0, 1, 4, 3]], [base[j] for j in [1, 2, 5, 4]],
        [base[j] for j in [2, 0, 3, 5]], [base[j] for j in [2, 1, 5, 4]]
    ]
    
    ax.add_collection3d(Poly3DCollection(faces, facecolors='cyan', linewidths=1, edgecolors='r', alpha=.5))

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
draw_centered_small_triangular_prism(ax)
ax.set_title('Üçgen Prizma')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
plt.show()
```

```python
def draw_centered_small_sphere(ax):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = 0.25 * np.outer(np.cos(u), np.sin(v))
    y = 0.25 * np.outer(np.sin(u), np.sin(v))
    z = 0.25 * np.outer(np.ones(np.size(u)), np.cos(v))

    ax.plot_surface(x, y, z, color='yellow', edgecolor='r')

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
draw_centered_small_sphere(ax)
ax.set_title('Küre')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
plt.show()
```

```python
def draw_centered_small_cone(ax, radius=0.25, height=0.5):
    theta = np.linspace(0, 2 * np.pi, 30)
    x = radius * np.cos(theta)
    y = radius * np.sin(theta)
    z = np.zeros_like(theta)
    
    top = np.array([x, y, height*np.ones_like(x)]).T
    bottom = np.array([x, y, np.zeros_like(x)]).T
    
    verts = [
        list(zip(bottom[:,0], bottom[:,1], bottom[:,2])),
        list(zip(top[:,0], top[:,1], top[:,2])),
        list(zip(np.concatenate((bottom[:,0], top[:,0])), 
                 np.concatenate((bottom[:,1], top[:,1])), 
                 np.concatenate((bottom[:,2], top[:,2]))))
    ]
    
    ax.add_collection3d(Poly3DCollection(verts, facecolors='orange', linewidths=1, edgecolors='r', alpha=.5))

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
draw_centered_small_cone(ax)
ax.set_title('Koni')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
plt.show()
```
